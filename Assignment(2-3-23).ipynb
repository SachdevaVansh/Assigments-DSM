It has module named pyplot which makes things easy for plotting by providing features to control plotting style, font properties, formatting axes etc.

#### Question 2-

A scatter plot helps in plotting a 2-D data using dots.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title")
plt.show()
```

#### Question 3-

The subplot() function is used to display multiple plots. 
The subplot() function takes three arguments that decribe the layout of the figure. The layout is organised in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,3,1) 
plt.plot(x,y)
   
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,3,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,3,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,3,4)
plt.plot(x,y)
```

#### Question 4-

A bar plot shows the relationship between a numeric and a categoric variable. These are used when we want to show segments of information i.e how much portion falls under which category.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.show()
```

```python
plt.barh(company,profit)
plt.show()
```

#### Question 5-

A box plot is the visual representation of the statistical five number summary of a given data set.
The five number summary is('minimum', 'first quatrtile', 'median','third quartile', and 'maximum').


These are used because they can summarize data from multiple sources and display the results in a single graph.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data)
plt.show()
```
