#### Question 1-

The different types of plots which can be plotted using seaborn are-
1. Line Plot
2. Count Plot
3. Bar plot
4. Pairplot
5. Scatterplot
6. Histogram


1. Lineplot- It is the simplest plot , as it is used to visualize only a single function. This plot helps to see the relation between x-axis and y-axis.

```python
import seaborn as sns
```

```python
x=[1,2,3]
y=[3,2,1]
sns.lineplot(x=x,y=y)
```

2. Countplot- A Count Plot Is Used To Show The Counts Of Observations In Each Categorical Bin Using Bars.

```python
iris=sns.load_dataset('iris')
sns.countplot(x='species', data=iris)
```

3. Bar Plot- A Bar Chart is a way of comparing a set of categorical Data.The Bar Chart Displays Data Using Several Bars, Each Representing A Particular Category.

```python
sns.barplot(x=iris['species'], y=iris['sepal_length'])
```

4. pair plot- Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column. 

```python
sns.pairplot(iris,hue='species')
```

5.Scatter plot- Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape. 

```python
import numpy as np
x=np.linspace(0,10,30)
y=np.sin(x)
sns.scatterplot(x=x,y=y,marker='s', color='blue')
```

6. Histogram-A Histogram Is A Graph That Shows The Underlying Frequency Distribution Of A Set Of Continuous Data. This Chart Is Helpful In Data Collection And Data Analysis And Hence It Is Widely Used In The Analytics Industry.

```python
x=[1,1,14,1,14,44,84,4,14,7,2,5]
y=[7,44,12,36,9,8,14,22,58,36,4,7]
sns.displot(x=x,y=y)
```
