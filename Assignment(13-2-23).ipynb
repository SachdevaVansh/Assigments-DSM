#### Question 3-

Three types of error are defined under ArithmeticError class-
1.ZeroDivisionError
2.OverflowError
3.FloatingpointError

#### ZeroDivisionError

```python
try :
    a=10/0
    logging.info(a)
except ZeroDivisionError as e:
    logging.info("Can't print due to {}".format(e)) 
```

#### OverflowError

```python
i=1
try:
    f=3.0**i
    for i in range(100):
        print(i,f)
        f=f**2
except OverflowError as err:
    print('Overflowed after',f,err)
```

#### Question 4-

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively. We can use LookupError exception class to handle both IndexError and KeyError exception classes.

```python
try:
    d={"key1":"dsm","key2":"python"}
    value=(d["key"])
    logging.info(value)
except KeyError as e:
    logging.info(e)
```

```python
try:
    l=[1,2,3,4,5]
    logging.info(l[5])
except IndexError as e:
    logging.info(e)
```
