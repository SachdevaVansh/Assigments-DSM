For discrete variables, the probability is straightforward and can be calculated easily. But for continuous variables which can take on infinite values, the probability also takes on a range of infinite values. The function which describes the probability for such variables is called a probability density function in statistics.

A function that defines the relationship between a random variable and its probability, such that you can find the probability of the variable using the function, is called a Probability Density Function (PDF) in statistics.

#### Question 3-

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt

# Creating a series of data of int type in range of 1-50

x=np.linspace(1,50,200)

# Creating the function
def normal_dist(x,mean,std):
    prob_density=(1/np.sqrt(2*np.pi)*std)*np.exp(-0.5*((x-mean)/std)**2)
    return prob_density

#Calculation of mean and standard deviation-
mean=np.mean(x)
std=np.std(x)

# Applying the function to the data
pdf=normal_dist(x,mean,std)

#Plotting the results
plt.plot(x,pdf, color='blue')
plt.xlabel("data points")
plt.ylabel("probability Density")
           
```

#### Question 4-

In statistics and probability theory, the binomial distribution is the probability distribution that is discrete and applicable to events having only two possible results in an experiment, either success or failure.
Properties-
1. There are only two distinct possible outcomes: true/false, success/failure, yes/no.
2. There is a fixed number of 'n' times repeated trials in a given experiment.
3. The probability of success or failure remains constant for each attempt/trial.
4. Only the successful attempts are calculated out of 'n' independent trials.
5. Every trial is an independent trial on its own, this means that the outcome of one trial has no effect on the outcome of another trial.

Example- Tossing a fair coin, Result of a student in an examination

#### Question 6-

```python
# Importing the required libraries
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt

# Creating a series of data of int type in range of 1-50

x=np.linspace(1,50,200)

# Creating the function
def poisson_dist(x,mean,t):
    pmf=(np.exp(-(mean/t))*(mean/t)**x)/(factorial(x))
    return pmf

#Calculation of mean and standard deviation-
mean=np.mean(x)

# Applying the function to the data
pmf=poisson_dist(x,mean,3)

#Plotting the results
plt.plot(x,pmf, color='red')
plt.xlabel("data points")
plt.ylabel("pmf")
           
```
